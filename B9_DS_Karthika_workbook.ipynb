```python
import pandas as pd
import numpy as np

df['Order_Date']=pd.to_datetime(df['Order_Date'])
df['Time_Orderd']=pd.to_datetime(df['Time_Orderd'])
df['Time_Order_picked']=pd.to_datetime(df['Time_Order_picked'])
df['Order_year']=df['Order_Date'].dt.year
df['Order_month']=df['Order_Date'].dt.month
df['Order_day']=df['Order_Date'].dt.day
df['Time_Orderd_hour']=df['Time_Orderd'].dt.hour
df['Time_Orderd_minutes']=df['Time_Orderd'].dt.minute
#df['Time_Orderd_seconds']=df['Time_Orderd'].dt.second
df['Time_Order_picked_hour']=df['Time_Order_picked'].dt.hour
df['Time_Order_picked_minutes']=df['Time_Order_picked'].dt.minute
#df['Time_Order_picked_seconds']=df['Time_Order_picked'].dt.second
df.drop(columns=['Order_Date','Time_Orderd','Time_Order_picked'],inplace=True)
df['Time_taken(min)']=df['Time_taken(min)'].str.lstrip('(min)')
df['Weatherconditions']=df['Weatherconditions'].str.lstrip('conditions')
#df['Time_taken(min)']=pd.to_numeric(df['Time_taken(min)'],errors='coerce').astype('int')
df['Time_taken(min)']=df['Time_taken(min)'].astype('int')

from sklearn import preprocessing
le=preprocessing.LabelEncoder()
for i in df.select_dtypes('object').columns:
  df[i]=le.fit_transform(df[i])
  r=6371
def deg_to_rad(degrees):
  return degrees*(np.pi/180)
def dist(lat1,lon1,lat2,lon2):
  lat=deg_to_rad(lat2-lat1)
  lon=deg_to_rad(lon2-lon1)
  a=np.sin(lat/2)**2 +np.cos(deg_to_rad(lat1))*np.cos(deg_to_rad(lat2))*np.sin(lon/2)**2
  c=2*np.arctan2(np.sqrt(a),np.sqrt(1-a))
  return r*c
for i in range(len(df)):
  df.loc[i,'distance']=dist(df.loc[i,'Restaurant_latitude'],
                            df.loc[i,'Restaurant_longitude'],
                            df.loc[i,'Delivery_location_latitude'],
                            df.loc[i,'Delivery_location_longitude'])

x=df.drop(columns=['Time_taken(min)','Restaurant_latitude','Restaurant_longitude','Delivery_location_latitude','Delivery_location_longitude'],axis=1)
y=df['Time_taken(min)']


from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x)
x=scaler.transform(x)
x=pd.DataFrame(x)

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)


from sklearn import linear_model
reg = linear_model.LinearRegression()
reg.fit(x_train, y_train)

y_pred = reg.predict(x_test)

from sklearn.metrics import r2_score
r2=r2_score(y_test,y_pred)
print('accuracy:',r2)
```

```python
from sklearn.neural_network import MLPRegressor
NN = MLPRegressor()
NN.fit(x_train, y_train)
y_pred = NN.predict(x_test)
print(np.vstack((y_test,y_pred)))

from sklearn.metrics import r2_score
r2=r2_score(y_test,y_pred)
print('accuracy:',r2)
```

```python
from sklearn.ensemble import RandomForestRegressor
NN = RandomForestRegressor()
NN.fit(x_train, y_train)
y_pred = NN.predict(x_test)
print(np.vstack((y_test,y_pred)))

from sklearn.metrics import r2_score
r2=r2_score(y_test,y_pred)
print('accuracy:',r2)
```
